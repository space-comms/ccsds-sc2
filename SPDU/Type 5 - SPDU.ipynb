```python
# Second Generation Lunar Use
class SPDU_type_5:
    def __init__(self,data,length):
        self.link_direction = None
        self.dir_function = None
        self.RNMD = None
        self.token = None
        self.mode = None
        self.polarisation = None
        self.coherent = None
        self.modulation = None
        self.mod_index = None
        self.coding = None
        self.ins_snr = None
        self.tranceiver_mode = None
        self.AOS_frame_length = None
        self.symbol_rate = None
        self.frequency = None

        self.PCID1 = None
        self.PCID0 = None
        self.time_tag_request = None
        self.status_report_request = None

        self.SEQ_CTRL_FSN = None

        self.SCID = None

        self.data = data
        self.directive = (data >> (length -3)) & 0x07
        self.select()

    def select(self):
        match self.directive:
            case 0x00:
                self.link_direction = (self.data >> 99) & 0x01
                self.dir_function = (self.data >> 96) & 0x07
                self.RNMD = (self.data >> 95) & 0x01
                self.token = (self.data >> 94) & 0x01
                self.mode = (self.data >> 91) & 0x07
                self.polarisation = (self.data >> 90) & 0x01
                self.coherent = (self.data >> 89) & 0x01
                self.modulation = (self.data >> 84) & 0x0F
                self.mod_index = (self.data >> 81) & 0x07
                self.coding = (self.data >> 75) & 0x1F
                self.ins_snr = (self.data >> 67) & 0xFF
                self.tranceiver_mode = (self.data >> 64) & 0x07
                self.AOS_frame_length = (self.data >> 48) & 0xFFFF
                self.symbol_rate = (self.data >> 32) & 0xFFFF
                self.frequency = self.data & 0xFFFFFFFF

            case 0x01:
                self.PCID1 = (self.data >> 12) & 0x01
                self.PCID0 = (self.data >> 11) & 0x01
                self.time_tag_request = (self.data >> 8) & 0x07
                self.status_report_request = (self.data >> 3) & 0x1F
            case 0x02:
                self.SEQ_CTRL_FSN = self.data & 0xFF
            case 0x03:
                self.SCID = self.data & 0xFFFF
```
