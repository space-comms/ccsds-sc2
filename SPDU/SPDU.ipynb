```python
# Directives/Reports for UHF operations
class SPDU_type_1:
    def __init__(self,data):
        self.data = data
        self.directive = (data >> 13)&0x07
        self.select()

        self.t_mode = None
        self.t_data_rate = None
        self.t_modulation = None
        self.t_encoding = None
        self.t_frequency = None

        self.time_sample = None
        self.duplex = None
        self.rmnd = None
        self.token = None

        self.r_mode = None
        self.r_data_rate = None
        self.r_modulation = None
        self.r_encoding = None
        self.r_frequency = None

        self.SEQ_CTRL_FSN = None

        self.status_report_request = None
        self.time_tag_request = None
        self.PCID0 = None
        self.PCID1 = None

        self.direction = None
        self.frequency_table = None
        self.rate_table = None
        self.carrier_mod = None
        self.data_mod = None
        self.mode_select = None
        self.scrambler = None
        self.Diff_Encoding = None
        self.RS_code = None

        self.SCID = None ##space id
    def select(self):
        match self.directive:
            case 0x00:
                self.t_frequency = (self.data >> 10) & 0x07
                self.t_encoding = (self.data >> 8) & 0x03
                self.t_modulation = (self.data >> 7) & 0x01
                self.t_data_rate = (self.data >> 3) & 0x07
                self.t_mode = self.data & 0x07
            case 0x01:
                self.token = (self.data >> 12) & 0x01
                self.rmnd = (self.data >> 11) & 0x01
                self.duplex = (self.data >> 6) & 0x07
                self.time_sample = (self.data >> 1) & 0x1F
            case 0x02:
                self.r_frequency = (self.data >> 10) & 0x07
                self.r_encoding = (self.data >> 8) & 0x03
                self.r_modulation = (self.data >> 7) & 0x01
                self.r_data_rate = (self.data >> 3) & 0x07
                self.r_mode = self.data & 0x07
            case 0x03:
                self.SEQ_CTRL_FSN = self.data & 0xFF
            case 0x04:
                self.PCID1 = (self.data >> 12) & 0x01
                self.PCID0 = (self.data >> 11) & 0x01
                self.time_tag_request = (self.data >> 8) & 0x07
                self.status_report_request = (self.data >> 3) & 0x1F
            case 0x06:
                self.RS_code = (self.data >> 12) & 0x01
                self.Diff_Encoding = (self.data >> 11) & 0x01
                self.scrambler = (self.data >> 9) & 0x03
                self.mode_select = (self.data >> 7) & 0x03
                self.data_mod = (self.data >> 5) & 0x03
                self.carrier_mod = (self.data >> 3) & 0x03
                self.rate_table = (self.data >> 2) & 0x01
                self.frequency_table = (self.data >> 1) & 0x01
                self.direction = self.data & 0x01
            case 0x07:
                self.SCID = self.data & 0x3FF
```
